```python
square_matrix = np.array([[4, 2],
                          [1, 3]])

print("\nSquare Matrix:")
print(square_matrix)

# Determinant
det = np.linalg.det(square_matrix)
print("\nDeterminant of Matrix:")
print(det)

# Inverse
inverse = np.linalg.inv(square_matrix)
print("\nInverse Matrix:")
print(inverse)

# Eigenvalues and Eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(square_matrix)

print("\nEigenvalues:")
print(eigenvalues)

print("\nEigenvectors:")
print(eigenvectors)

tensor = np.arange(24).reshape(2, 3, 4)

print("\n3D Array (Tensor):")
print(tensor)

print("\nShape of Tensor:")
print(tensor.shape)

print("\nNumber of Dimensions:")
print(tensor.ndim)

print("\nElement at Position [1,2,3]:")
print(tensor[1, 2, 3])

size = 1000000

x = np.arange(size)
y = np.arange(size)

# Vectorized Addition
start = time.time()

result_vectorized = x + y

vectorized_time = time.time() - start

print("\nVectorized Addition Time:")
print(vectorized_time, "seconds")


# Loop-Based Addition
start = time.time()

result_loop = np.zeros(size)

for i in range(size):
    result_loop[i] = x[i] + y[i]

loop_time = time.time() - start

print("\nLoop-Based Addition Time:")
print(loop_time, "seconds")


# Speed Comparison
speedup = loop_time / vectorized_time

print("\nSpeedup Factor:")
print(speedup)

print("\nMean of Tensor:")
print(np.mean(tensor))

print("\nMaximum Value in Tensor:")
print(np.max(tensor))

print("\nSum of All Elements in Tensor:")
print(np.sum(tensor))
```
