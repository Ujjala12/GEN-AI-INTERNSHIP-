```python
import pandas as pd

# Load CSV dataset
df = pd.read_csv("cleaned_employees.csv")

# Display first five rows
print(df.head())
```

```python
# Dataset information
print(df.info())

# Check for missing values
print(df.isnull().sum())

# Summary statistics
print(df.describe())
```

```python
# Fill missing salary with mean
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())

# Fill missing age with median
df["Age"] = df["Age"].fillna(df["Age"].median())

# Fill missing department with mode
df["Department"] = df["Department"].fillna(df["Department"].mode()[0])
```

```python
df["Joining_Date"] = pd.to_datetime(df["Joining_Date"])
```

```python
avg_salary = df.groupby("Department")["Salary"].mean()

print(avg_salary)
```

```python
summary = df.groupby("Department").agg({
    "Salary": ["mean", "max", "min"],
    "Age": "mean",
    "Employee_ID": "count"
})

print(summary)
```

```python
import pandas as pd
df = pd.read_csv("cleaned_employees.csv")

print("========== EMPLOYEE DATASET ==========")
print(df)

rows, cols = df.shape

print("\nTotal Rows :", rows)
print("Total Columns :", cols)

missing = df.isnull().sum()

print("\nMissing Values")
print(missing)

# Logic to handle missing values
if missing.sum() > 0:
    print("\nMissing values found!")

    # Numerical columns
    df["Salary"].fillna(df["Salary"].mean(), inplace=True)
    df["Age"].fillna(df["Age"].median(), inplace=True)

    # Categorical column
    df["Department"].fillna(df["Department"].mode()[0], inplace=True)

    print("Missing values filled successfully.")
else:
    print("No missing values found.")

duplicate = df.duplicated().sum()

if duplicate > 0:
    print("\nDuplicate Records Found :", duplicate)
    df.drop_duplicates(inplace=True)
    print("Duplicates removed.")
else:
    print("\nNo duplicate records found.")

df["Joining_Date"] = pd.to_datetime(df["Joining_Date"])

bonus = []

for salary in df["Salary"]:

    if salary >= 80000:
        bonus.append(salary * 0.20)

    elif salary >= 60000:
        bonus.append(salary * 0.15)

    else:
        bonus.append(salary * 0.10)

df["Bonus"] = bonus

# Total Salary
df["TotalSalary"] = df["Salary"] + df["Bonus"]


grade = []

for salary in df["Salary"]:

    if salary >= 90000:
        grade.append("A")

    elif salary >= 70000:
        grade.append("B")

    elif salary >= 50000:
        grade.append("C")

    else:
        grade.append("D")

df["Grade"] = grade

current_year = 2026

experience = []

for date in df["Joining_Date"]:
    experience.append(current_year - date.year)

df["Experience"] = experience


print("\nEmployees with Salary > 60000")

high_salary = df[df["Salary"] > 60000]

print(high_salary)


summary = df.groupby("Department").agg({

    "Salary":["mean","max","min"],

    "Bonus":"sum",

    "Employee_ID":"count"

})

print("\nDepartment Summary")
print(summary)


highest = df.loc[df["Salary"].idxmax()]

print("\nHighest Paid Employee")
print(highest)


lowest = df.loc[df["Salary"].idxmin()]

print("\nLowest Paid Employee")
print(lowest)


df.to_csv("cleaned_dataset.csv", index=False)

summary.to_csv("department_summary.csv")


print("\n")
print("="*60)
print("         CLEANED DATASET REPORT")
print("="*60)

print("Dataset Loaded Successfully")
print("Rows :", len(df))
print("Columns :", len(df.columns))

print("\nCleaning Performed")
print("----------------------")

if missing.sum() == 0:
    print("✓ No Missing Values")
else:
    print("✓ Missing Values Filled")

if duplicate == 0:
    print("✓ No Duplicate Records")
else:
    print("✓ Duplicate Records Removed")

print("✓ Joining Date Converted")
print("✓ Bonus Calculated")
print("✓ Salary Grade Assigned")
print("✓ Experience Calculated")

print("\nDepartment-wise Aggregation Completed")

print("\nDataset Saved Successfully")

print("="*60)
```
