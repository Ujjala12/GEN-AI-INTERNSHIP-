```python
Q1 = df["amount"].quantile(0.25)
Q3 = df["amount"].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

outliers = df[
    (df["amount"] < lower_limit) |
    (df["amount"] > upper_limit)
]

print("Outliers")
print(outliers)
```

```python
df_clean = df[
    (df["amount"] >= lower_limit) &
    (df["amount"] <= upper_limit)
]
```

```python
df_clean.to_csv(
    "budget_cleaned.csv",
    index=False
)
```
